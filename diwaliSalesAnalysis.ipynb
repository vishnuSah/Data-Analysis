```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Diwali Sales Data.csv", encoding='latin')
df.head()
```

```python
df.info()
```

```python
df.drop(['Status','unnamed1'], axis=1, inplace=True)
```

```python
df.info()
```

```python
pd.isnull(df).sum()
```

```python
df.dropna(inplace=True)
df.isnull().sum()
```

```python
df['Amount'] = df['Amount'].astype('int64')
```

```python
df.describe()
```

```python
df[['Age','Orders','Amount']].describe()
```

Exploratory Data Analysis (EDA)

```python
df.head()
```

Gender

```python
plt.figure(figsize=(6,5))
ax = sns.countplot(data=df, x='Gender')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df.head()
```

```python
gbGender = df.groupby('Gender').agg({'Amount':'sum'}).sort_values(by='Amount', ascending=True)
#plt.figure(figsize=(8,7))
ax = sns.barplot(x='Gender', y='Amount',hue='Gender', data=gbGender, err_kws={'linewidth': 0})
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
gbGender.reset_index()
```

From above two chart we conclude that Womens likes more to shopping and the total amount spent by females are more compared to male.

Age

```python
df.head()
```

```python
sortedDF =df.sort_values(by='Age Group')
```

```python
ax = sns.countplot(x='Age Group', data=sortedDF, hue='Gender')
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
gbAge = df.groupby(['Age Group','Gender'], as_index=False).agg({'Amount':'sum'})#.reset_index()
gbAge = gbAge.assign(Amt = lambda x: round(x['Amount']/100000,2))
gbAge
```

```python
plt.figure(figsize=(8,6))
ax = sns.barplot(x='Age Group', y='Amt',data=gbAge, hue='Gender', err_kws={'linewidth': 0})
ax.set(ylabel='Amount in Lakhs')
plt.show()
```

From above chart we can see that the count of female shopper is more and mount spent by females is also higher compared to males

State

```python
df.head()
```

```python
gbState = df.groupby(['State'], as_index=False).agg({'Orders':'sum'}).sort_values(by='Orders',ascending=False).head(10)
gbState
```

```python
plt.figure(figsize=(15,5))
ax = sns.barplot(x='State', y='Orders', data=gbState, hue='State')
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

Occupations

```python
df.head()
```

```python
df['Occupation'].value_counts()
```

```python
plt.figure(figsize=(18,5))
ax=sns.countplot(x='Occupation', data=df, hue='Occupation')
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

```python
gbOccupation = df.groupby(['Occupation'], as_index=False).agg({'Amount':'sum'}).sort_values(by='Amount', ascending=False).head()
gbOccupation = gbOccupation.assign(Amt= lambda x:round(x['Amount']/100000,2))
gbOccupation
```

```python
plt.figure(figsize=(8,5))
ax=sns.barplot(x='Occupation', y='Amt',data=gbOccupation, hue='Occupation')
ax.set(ylabel='Amount (in Lakhs)')
for bar in ax.containers:
    ax.bar_label(bar)
plt.show()
```

From Above data and chart we conclude that people working in IT sector has done more shopping smilarly list of other top 4 spending as per Occupation.

Product 

```python
df.head()
```

```python
gbProduct = df.groupby(['Product_Category'], as_index=False).agg({'Orders':'sum'}).sort_values(by='Orders', ascending=False).head(8)
gbProduct
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x='Product_Category', y='Orders', data=gbProduct)
for bar in ax.containers:
    ax.bar_label(bar)
```

```python
gbProductAmt = df.groupby(['Product_Category'], as_index=False).agg({'Amount':'sum'}).head(10)
gbProductAmt = gbProductAmt.assign(Amt= lambda x: round(x['Amount']/10000,2))
gbProductAmt
```

```python
plt.figure(figsize=(15,6))
ax = sns.barplot(x='Product_Category', y='Amt', data=gbProductAmt, hue='Product_Category')
ax.set(ylabel='Amount (in Lakhs)')
ax.set(xlabel='Products Category')
for bar in ax.containers:
    ax.bar_label(bar)
```

From above chart we can see that Clothing, Food, Electronic are most sold product but amount spent on Food is way higher than all categories.

Conclusion:

From above analysis we conclude that womens of age between 26 to 35 living in UP, Maharashtra or Karnataka working in IT, healthcare, Aviation spent more on clothing, food and electronic gadgets.
