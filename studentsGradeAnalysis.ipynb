```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sea
```

```python
df = pd.read_csv('StudentsPerformance.csv')
df.head()
```

```python
df.count()
```

```python
plt.figure(figsize=(5,4))
ax = sea.countplot(x='gender', data=df)
ax.bar_label(ax.containers[0])
```

From above chart we can observe that number of female students is greater than male students.

```python
gbs = df.groupby('parental level of education').agg({'math score':'mean','reading score':'mean', 'writing score':'mean'})
gbs
```

```python
plt.figure(figsize=(6,4))
sea.heatmap(gbs, annot=True)
plt.title('Relation of parents education and students Marks')
plt.show()
```

From above chart we conclude that parents education have impact on their childrens marks 

```python
gbs1 = df.groupby('race/ethnicity').agg({'math score':'mean','reading score':'mean','writing score':'mean'})
gbs1
```

```python
ax = sea.countplot(data=df, x='ethnicity')
ax.bar_label(ax.containers[0])
plt.show()
```

```python
df.rename(columns={'race/ethnicity':'ethnicity'}, inplace=True)
df.head()
```

```python
groupA = df.loc[(df['ethnicity'] == 'group A')].count()
groupB = df.loc[(df['ethnicity'] == 'group B')].count()
groupC = df.loc[(df['ethnicity'] == 'group C')].count()
groupD = df.loc[(df['ethnicity'] == 'group D')].count()
groupE = df.loc[(df['ethnicity'] == 'group E')].count()
#groupB['ethnicity']

l = ['group A','group B','group C','group D','group E']
mlist = [groupA['ethnicity'], groupB['ethnicity'], groupC['ethnicity'], groupD['ethnicity'], groupE['ethnicity']]
plt.pie(mlist, labels=l, autopct='%1.2f%%')
plt.title('Distribution of Ethnic Group')
plt.show()
```
